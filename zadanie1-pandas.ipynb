```python
import pandas as pd
data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/', header=0)
```

```python
data
```

```python
df = data[0]
df
```

<b>Zamień nagłówki kolumn na polskie odpowiedniki: ['TYTUŁ','ARTYSTA','ROK','MAX POZ']
</b>

```python
df.rename(columns={'POS':'LP', 'TITLE':'TYTUŁ','ARTIST':'ARTYSTA','YEAR':'ROK','HIGH POSN':'MAX POZ'}, inplace=True)
```

```python
df.iloc[0]
```

<b>Ilu pojedynczych artystów znajduje się na liście?</b>

```python
df['ARTYSTA'].nunique()
```

<b>Które zespoły pojawiają się najczęściej na liście?</b>

```python
df['ARTYSTA'].value_counts().head(2)
```

Najczęściej pojawiające się zespoły to: Coldplay i Take That.

<b>Zmień nagłówki kolumn, tak aby każdy z nich rozpoczynał się od wielkiej litery, a pozostałe były wprowadzone małymi literami.</b>

```python
df.rename(str.capitalize, axis='columns', inplace=True)
df.head()
```

<b>Wyrzuć z tabeli kolumnę ‘Max poz’.</b>

```python
df.drop('Max poz', axis=1, inplace=True)
```

```python
df.head()
```

<b>W którym roku wyszło najwięcej albumów znajdujących się na liście?</b>

```python
# df.groupby('Rok').count().sort_values(by='Tytuł', ascending=False).head(2)
df['Rok'].value_counts().head(2)
```

W latach 2000 i 1987 wyszło najwięcej albumów (po 4 albumy).

<b>Ile albumów wydanych między 1960 a 1990 rokiem włącznie znajduje się na liście?</b>

```python
number_albums_from_specific_period = df[(df['Rok']>1960) & (df['Rok']<=1990)]['Lp']
len(number_albums_from_specific_period)
```

Na liście znajdują się 22 albumy z tego okresu.

<b>W którym roku wydany został najmłodszy album na liście?</b>

```python
df_sort_desc = df.sort_values(by='Rok', ascending=False)
df_sort_desc.iloc[0]
```

Najmłodszy album został wydany w 2015 r.

<b>Przygotuj listę najwcześniej wydanych albumów każdego artysty, który znalazł się na liście</b>

```python
list_album = df.sort_values(by='Rok')
list_album.drop_duplicates(subset=['Artysta'], inplace=True)
list_album.groupby(['Artysta', 'Tytuł', 'Rok']).count().sort_values(by='Rok')
list_album.drop('Lp',axis=1, inplace=True)
list_album
```

<b>Listę zapisz do pliku csv.</b>

```python
list_album.to_csv('list_album.csv', index=False)
```
